```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
import seaborn as sns
```

```python
# load first all date set and append
df1 = pd.read_excel('Superstore.xlsx', sheet_name='Central')
df2 = pd.read_excel('Superstore.xlsx', sheet_name='East')
df3 = pd.read_excel('Superstore+South.xlsx', sheet_name='South')
df4 = pd.read_excel('Superstore.xlsx', sheet_name='West')
```

```python
# append all table into a single table
df = pd.concat([df1,df2,df3,df4])
```

```python
# show the first 5 rows
df.head()
```

```python
# retrieve the last 5 rows
df.tail(5)
```

```python
# check for null values
df.isna()
```

```python
#check for data information
df.info()
```

```python
# get the statistic info
df.select_dtypes(include='float').describe()
```

```python
# create a new column by calculating the total sales
df["TOTAL SALES"] = df["SALES"] * df["QUANTITY"]
df.head()
```

```python
df.shape
```

```python
# delete ROW ID column
del df['ROWID']
```

```python
# add a new column and name it row id, then set it as index
df['ROLLID'] = np.arange(1,9995)
df = df.set_index('ROLLID')
df.head(1)
```

```python
#create a new column by calculating profit %
df["PROFIT RATE"] = df["PROFIT"] / df["TOTAL SALES"]
df.head()
```

```python
# check column names. change any if necessary
df.columns
```

```python
df.columns = ['ORDER ID', 'ORDER DATE', 'SHIP DATE', 'SHIP MODE', 'CUSTOMER ID',
       'CUSTOMER NAME', 'SEGMENT', 'COUNTRY', 'CITY', 'STATE', 'POST CODE',
       'REGION', 'PRODUCT ID', 'CATEGORY', 'SUB CATEGORY', 'PRODUCT NAME',
       'SALES', 'QUANTITY', 'DISCOUNT', 'PROFIT', 'TOTAL SALES',
       'PROFIT RATE']
```

```python
#capitalise all column names
df.columns = df.columns.str.capitalize()
df.head(1)
```

```python
# calculate the total sales that was generated
total_sales = df['Total sales'].sum()
round(total_sales,2)
```

### calculate the total profit and sales made in each region.

```python
#calculate the total profit and sales made in each region.

pro_sales_total = df.groupby('Region')['Profit','Total sales'].agg(sum)

pro_sales_total 
```

### Create a visulization for total sales made in every region

```python
# regional sales visualization
reg_sales = df.groupby('Region')['Total sales'].agg(sum)
label = df['Region'].unique()

fig1, ax1 = plt.subplots()
ax1.pie(reg_sales,  labels=label, autopct='%1.1f%%')
ax1.axis('equal') 
plt.show()
```

### show the top 10 customers by total profit


```python
top_ten = df.groupby('Customer name')['Profit'].sum().sort_values(ascending=False).head(10) 
print(top_ten)

#top 10 visualization
top_ten.plot(kind='bar', figsize=(12,6), color='green', 
             title='Top 10 Customer by Total Profit', ylabel='Profit'
            )
```

### show the bottom 10 customers by profit

```python
bottom_ten = df.groupby('Customer name')['Profit'].sum().sort_values().head(10) 
print(bottom_ten)


```

```python
#top 10 visualization
bottom_ten.plot(kind='barh', figsize=(12,6), color='red', 
             title='Bottom 10 Customer by Total Profit', ylabel='Profit'
            )
```

### show the yearly trend by sales and profit

```python
#yearly sales
yearly_sales = df.pivot_table(index='Year', values='Profit')
yearly_sales
```

```python
#yearly profit
yearly_Profit = df.pivot_table(index='Year', values='Profit')
yearly_Profit
```

```python
# yearly sales and profit trend visualization

fig,axes =  plt.subplots(nrows=2,ncols=1, figsize=(12,8))


axes[0].plot(yearly_sales, color='r',
                  lw=2, ls='--', marker = 'o',
                  markersize=20,  markerfacecolor='green',
                  markeredgecolor='yellow',markeredgewidth=3)

axes[0].set_title('Yearly Trend by Sales',fontsize=22)
axes[0].set_xlabel('Year',fontsize=20)
axes[0].set_ylabel('Sales',fontsize=20)


axes[1].plot(yearly_Profit,color='blue',
                  lw=5, alpha=0.5, marker = '*',
                  markersize=20,  markerfacecolor='red',
                  markeredgecolor='black',
                  markeredgewidth=3)
axes[1].set_title('Yearly Trend by Profut',fontsize=22)
axes[1].set_xlabel('Year',fontsize=16)
axes[1].set_ylabel('Profit',fontsize=16)

plt.tight_layout()
```
