```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
import seaborn as sns
```

```python
# load first all date set and append
df1 = pd.read_excel('store.xlsx', sheet_name='Central')
df2 = pd.read_excel('store.xlsx', sheet_name='East')
df3 = pd.read_excel('store.xlsx', sheet_name='South')
df4 = pd.read_excel('store.xlsx', sheet_name='West')

```

```python
# append all table into a single table
df = pd.concat([df1,df2,df3,df4])
```

```python
# change all column to upper case
df.columns = df.columns.str.upper()
df.columns
```

```python
# create a new column by calculating the total sales
df["TOTAL SALES"] = df["SALES"] * df["QUANTITY"]
df.head()
```

```python
#create a new column by calculating profit %
df["PROFIT RATE"] = df["PROFIT"] / df["TOTAL SALES"]
df.head()
```

```python
# calculate the total sales made
total_sales = df["TOTAL SALES"].sum()
round(total_sales ,2)
```

```python
# calculate the average sales
Avg_sales = df["TOTAL SALES"].mean()
round(Avg_sales,2)
```

```python
# calculate the total profit made

total_profit = df["PROFIT"].sum()
round(total_profit ,2)
```

#  calculate the total profit and sales made in each region.


```python
regional_sales_profit = df.groupby('REGION')['TOTAL SALES', 'PROFIT'].agg(sum)
print(regional_sales_profit)


# regional sales chart
reg_sales = df.pivot_table(index='REGION',values='TOTAL SALES',
                           aggfunc=sum)
reg_sales.plot(kind='pie',y='TOTAL SALES',title='Regional Sales', 
               legend=False,) 



# regional profit chart 
reg_profit = df.pivot_table(index='REGION',values='PROFIT',
                           aggfunc=sum)
reg_profit.plot(kind='pie',y='PROFIT',title='Regional Profit', 
                legend=False,) 
```

# show the top 10 customers by total profit

```python
top_ten = df.groupby('CUSTOMER NAME')['PROFIT'].agg(sum)
top_ten_customers = top_ten.sort_values(ascending=False).head(10)
print(top_ten_customers)

top_ten_customers.plot.bar(figsize=(12,5), color='red',
                          ylabel='Profit',
                           title='Top 10 Customers by Profit', 
                            fontsize=12
                          )
```

# show the bottom 10 customers by profit

```python
bottom_ten = df.groupby('CUSTOMER NAME')['PROFIT'].agg(sum)
bottom_ten_customers = bottom_ten.sort_values().head(10)
print(bottom_ten_customers) 

bottom_ten_customers.plot.barh(
figsize=(8,7), color='orange', ylabel='Profit', 
    title='Bottom 10 Customers by Profit'
                               )
```

# show the yearly trend by sales and profit

```python
yearly_sales = df.groupby('YEAR')['TOTAL SALES'].agg(sum)
print(yearly_sales)



yearly_profit = df.groupby('YEAR')['PROFIT'].agg(sum)
print(yearly_profit)


fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(12,7))

axes[0].plot(yearly_sales,  color='r',
                  lw=2, ls='--', marker = 'o',
                  markersize=20,  markerfacecolor='green',
                  markeredgecolor='yellow',
                  markeredgewidth=3)
axes[0].set_title('Yearly Trend by Sales',fontsize=22)
axes[0].set_xlabel('Year',fontsize=20)
axes[0].set_ylabel('Sales',fontsize=20)



axes[1].plot(yearly_profit,  color='blue',
                  lw=5, alpha=0.5, marker = '*',
                  markersize=20,  markerfacecolor='red',
                  markeredgecolor='black',
                  markeredgewidth=3)
axes[1].set_title('Yearly Trend by Profit',fontsize=22)
axes[1].set_xlabel('Year',fontsize=16)
axes[1].set_ylabel('Profit',fontsize=16)

plt.tight_layout()
```
