```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the data set you want to join
customer_Info = pd.read_excel('CustomerOrders-01.xlsx',sheet_name='Customer Info')
Order_Info = pd.read_excel('CustomerOrders-01.xlsx',sheet_name='Order Info')
```

```python
# retrieve customer info first 5 rows
customer_Info.head()
```

```python
# retrieve customer info first 5 rows
Order_Info.head()
```

```python
# rename order info column Ship name to Company name to be able to join both data set
Order_Info = Order_Info.rename(columns={'ShipName':'Company Name'})
Order_Info.columns
```

```python
# join the table to retrive matching records
pd.merge(customer_Info,Order_Info,on='Company Name')
```

```python
# left join both data set to retrive left matching values
pd.merge(customer_Info,Order_Info,on='Company Name', how='left')
```

```python
# right join both data set to retrive rightt matching values
pd.merge(customer_Info,Order_Info,on='Company Name', how='right')
```

```python
# retrieve all values 
pd.merge(customer_Info,Order_Info,on='Company Name', how='outer')
```
