### PIVOT TABLE

```python
data.head(2)
```

```python
# Sum up the each regional profit
Region_profit = data.pivot_table(
    index = 'REGION',
    values = 'PROFIT',
    aggfunc = sum
)
print(Region_profit)

# pie chart for Region_profit
Region_profit.plot(kind='pie', y='PROFIT', title='Regional Profit')
```

```python
#top 5 customers
Top_5_customer = data.pivot_table(
    index='CUSTOMER NAME',
    values='PROFIT',
    aggfunc=sum
)
top_cus= Top_5_customer.sort_values('PROFIT',ascending=False).head()
print(top_cus)

# top 5 profitable customers
top_cus.plot(kind='bar', ylabel='Profit', title='Top 5 Customers by Profit', figsize=(10,5))
```

```python
#bottom 5 customers
bottom_5_customer = data.pivot_table(
    index='CUSTOMER NAME',
    values='PROFIT',
    aggfunc=sum
)
bottom5 = bottom_5_customer.sort_values('PROFIT').head()
bottom5
```

```python
# top 5 profitable states
top_5_states = data.pivot_table(index='STATE',
                               values='PROFIT',
                               aggfunc=sum)
top5 = top_5_states.sort_values('PROFIT', ascending=False).head()
print(top5)

# top 5 profitable states
top5.plot(kind='barh', title='Top 5 Profitable States', color ='orange', ylabel='Profit', xlabel='Profit',
          
 figsize=(8,6))
```

```python
# bottom 5 profitable states
bottom_5_states = data.pivot_table(index='STATE',
                               values='PROFIT',
                               aggfunc=sum)
bottom5 = bottom_5_states.sort_values('PROFIT').head()
bottom5
```

```python
# month/year profit trend

my = data.pivot_table(index='MONTH', columns = 'YEAR', values='PROFIT', aggfunc=sum)
my
```

```python
# calculate the profit by category
category_profit_y = data.pivot_table(index='YEAR',columns='CATEGORY', values='PROFIT', aggfunc=sum)
print(category_profit_y)

#  yearly profit by category
category_profit_y.plot(kind='line', title= 'Yearly Profit by Category', ylabel='Profit', figsize=(10,5))
```

```python
# calculate profit made yearly by sub category
sub_cat_y = data.pivot_table(index='SUB-CATEGORY',columns='YEAR', values='PROFIT', aggfunc=sum)
sub_cat_y 
```

```python
# calculate total profit by segment
segment = data.pivot_table(index='YEAR', columns='SEGMENT', values='PROFIT', aggfunc=sum)
segment

# profit by segment

segment.plot(kind='bar', title='Profit by Segement', figsize=(10,5))
```
