```python
import pandas as pd
import numpy as np

```

```python
df=pd.read_excel('Population Report.xlsx', sheet_name='Employee')
df
```

```python
# check type types and see if it is necessary to change some of the column data type
df.dtypes
```

```python
# convert salary data type to float
df=df.astype({'Salary':int})
df.dtypes
```

```python
# now reshape the data frame to show total salary paid to each department
df_salary=df.pivot_table(
index='Department',
    values='Salary',
    aggfunc="sum"
)
df_salary
```

```python
# which dept has the highest total salary
highest=df_salary.sort_values(by='Salary', ascending=False)

highest.head(1)
```

```python
## which dept has the lowest total salary
lowest=df_salary.sort_values(by='Salary', ascending=True)
lowest.head(1)
```

```python
# or
low_salary=df_salary.sort_values(by='Salary', ascending=False)

low_salary.tail(1)
```

```python
# check the salary paid to each department in every state
state_salary=df.pivot_table(
columns='Department',
    index='State of Origin',
    values='Salary',
    aggfunc="sum"
)

state_salary
```

```python
# retrive the total salary and bonus paid to each state
state_salary_bonus=df.pivot_table(

    index='State of Origin',
    values={'Salary','Departmental Bonus'},
    aggfunc="sum"
)

state_salary_bonus
```

```python
# select the top 5 states that received highest bonus
```

```python
highest_state_salary=df.pivot_table(
index='State of Origin',
values='Departmental Bonus',
    aggfunc="sum"

)
highest=highest_state_salary.sort_values('Departmental Bonus', ascending=False)
highest.head(5)
```

```python
# lowest 5 total salary

highest.tail(5)


```

```python
# which employees are overall top 5 salary earners 
top_earners=df.pivot_table(
index='Employee Names',
values='Salary',
    aggfunc="sum"

)
top_earners.sort_values('Salary', ascending=False).head(5)
```

```python
# which employees are overall bottom 5 salary earners 
top_earners=df.pivot_table(
index='Employee Names',
values='Salary',
    aggfunc="sum"

)
top_earners.sort_values('Salary', ascending=True).head(5)
```

```python
# check the top 5 earner including their location
o=df.pivot_table(
    index=['State of Origin','Employee Names'],
values='Salary',
    aggfunc="sum")
o.sort_values('Salary', ascending=False).head(5)
```

```python
# lowest including their state
o.sort_values('Salary', ascending=True).head(5)
```

```python
# retrive the top 5 earners department
dpt=df.pivot_table(
    index=['Department','Employee Names'],
values='Salary',
    aggfunc="sum")
dpt.sort_values('Salary', ascending=False).head(5)
```

```python
dpt_lowest=df.pivot_table(
    index=['Department','Employee Names'],
values='Salary',
    aggfunc="sum")
dpt_lowest.sort_values('Salary', ascending=True).head(5)
```
