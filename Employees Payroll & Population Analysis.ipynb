### PAYROLL QUESTIONS 
B
1. Creata a new column and calculate the total salary paid by summing up salary and bonus
2. list the top 10 salary earners.
3. list the bottom 10 salary earners.
4. Calculate the total salary, average salary, total bonus paid to every employees on each location.
5. find the total salary, average salary and bonus paid to each department.
6. what is the total salary paid to all workers
7. what is the totalincome (salary + bonus) to all workers.
8. what is the total bonus paid to all workers.

### TASK B SOLUTIONS

```python
# 1. Creata a new column and calculate the total salary paid by summing        up salary and bonus
df['TOTAL INCOME'] = df['SALARY'] + df['BONUS']
df.head()
```

```python
#  2.list the top 10 salary earners.
bottom_ten_earners = df.sort_values(by='SALARY', ascending=False)
bottom_ten_earners
```

```python
#  3.list the bottom 10 salary earners.
bottom_ten_earners = df.sort_values(by='SALARY')
bottom_ten_earners
```

```python
# 4. Calculate the total salary, average salary, total bonus paid to every employees on each location.
location_salary = df.groupby('LOCATION')[['SALARY','BONUS']].agg([sum,np.mean])
location_salary 
```

```python

# 5. find the total salary, average salary and bonus paid to each department.
dept_salary = df.groupby('DEPARTMENT')[['SALARY','BONUS']].agg([sum,np.mean])
dept_salary

```

```python
6. # what is the total salary paid to all workers
tot_salary = df['SALARY'].sum()
tot_salary
```

```python

# 7. what is the total income (salary + bonus_ to all workers.

total_income = df['TOTAL INCOME'].sum()
total_income
```

```python
# 8. what is the total bonus padi to all workers.
total_bonus = df['BONUS'].sum()
total_bonus
```

```python
# extract the year, month, day of the week from the start date column
df['YEAR'] = df['START DATE'].dt.year
df['MONTH'] = df['START DATE'].dt.month
df['DAY NAME'] = df['START DATE'].dt.day_name()
```

```python

df.head(2)
```

```python
# create a pie chart to show the total salary paid to both male and female workers
Gender_pivot = df.pivot_table(index='GENDER', values = 'SALARY',
                             aggfunc = sum)
Gender_pivot
```

```python
Gender_pivot.plot(kind='pie', y = 'SALARY')
```

```python
# create a bar chat to show the total salary paid to all department

dept_salary = df.pivot_table (index='DEPARTMENT',
                            values='SALARY',
                            aggfunc = sum)

dept_salary 
```

```python

dept_salary.plot(kind='bar', ylabel = 'TOTAL SALARY',
                title = 'DEPARTMENTAL SALARY', color='orange')
```

```python
# create a bar chat showing the total salary paid to both male and female in every department

f_m = df.pivot_table(index = 'DEPARTMENT', columns='GENDER', 
                     values='SALARY',
                    aggfunc=sum)
f_m 
```

```python
f_m.plot(kind='bar', title = 'SALARY PAID PER GENDER IN DEPARTMENT',
        figsize=(12,4))
```

```python
# create a line chart to show total income (i.e salary and bonus) paid every year
year_salary = df.pivot_table(index='YEAR', 
                               columns='GENDER',
                               values = 'SALARY',
                               aggfunc= sum)
year_salary
```

```python
# check only 1998 departmental salary
year1998=year_salary[year_salary.index.isin([1998])]
year1998 = year1998.T
year1998 
```

```python
year1998.plot(kind='bar', title =' 1998 TOTAL SALARY ')
```
